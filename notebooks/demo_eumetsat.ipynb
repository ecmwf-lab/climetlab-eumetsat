```python
from climetlab import load_source, plot_map

ds = load_source("eumetsat-datastore", "EO:EUM:DAT:METOP:GLB-SST-NC")

plot_map(
    ds,
    style=dict(
        contour=False,
        contour_shade=True,
        contour_shade_colour_direction="clockwise",
        contour_shade_max_level_colour="magenta",
        contour_shade_min_level_colour="cyan",
        contour_shade_technique="cell_shading",
        contour_shade_cell_resolution=100,
    ),
)
```
